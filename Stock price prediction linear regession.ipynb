```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
```

```python
pd.set_option("display.max_rows",None)
pd.set_option("display.max_column",None)
```

```python
df = yf.download("M&M.NS", period = "ytd",
    progress = False
)


display(df)

df.Close.plot()
```

```python
df = yf.download(
    "TM",
    period = "ytd",
    progress = False
)


display(df)

df.Close.plot()
```

# download data to check the valuation every minute

```python
df = yf.download(
    "TSLA",
    period = "7d",
    interval = "1m",
    progress = False
)


display(df)

df.Close.plot()
```

```python
# multiple ticker
df = yf.download(
    ["M&M.NS", "TSLA", "TM"],
    start = "2020-01-01",
    end = "2023-06-14",
    progress = False
)


display(df)

df.Close.plot()
```

```python
df = yf.download(
    ["M&M.NS", "TSLA", "TM"],
    start = "2020-01-01",
    end = "2023-06-14",
    group_by = "ticker",
    progress = False
)


display(df)
```
